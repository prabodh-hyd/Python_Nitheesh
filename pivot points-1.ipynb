# pivot points

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_file_path = "D:\\Prabodh\\python For Trading\\nifty50_historical_data.csv"

df = pd.read_csv(csv_file_path)

df = df.dropna()

df['Date'] = pd.to_datetime(df['Date'])

high = df['High'].values
low = df['Low'].values
close = df['Close'].values

pivot_points = (high + low + close) / 3
bottom_central = (high + low) / 2
top_central = pivot_points + (pivot_points - bottom_central)

resistance_1 = (2*pivot_points-low)
support_1 = (2*pivot_points - high)

resistance_2 = (pivot_points+ (high - low))
support_2 = (pivot_points - (high - low))

resistance_3 = (high + 2*(pivot_points - low))
support_3 = (low - 2*(high - pivot_points))

df['Pivot Points'] = pivot_points
df['Bottom Central'] = bottom_central
df['Top Central'] = top_central
df['Resistance-1'] = resistance_1
df['Support-1'] = support_1
df['Resistance-2'] = resistance_2
df['Support-2'] = support_2
df['Resistance-3'] = resistance_3
df['Support-3'] = support_3 

df.to_csv("D:\\Prabodh\\python For Trading\\nifty50_pivot_points.csv", index = False)
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', linestyle = 'solid', color='black')
plt.plot(df['Date'], df['Pivot Points'], label='Pivot Points', linestyle='dashed', color='blue')
plt.plot(df['Date'], df['Resistance-1'], label='Resistance 1', linestyle='dotted', color='red')
plt.plot(df['Date'], df['Support-1'], label='Support 1', linestyle='dotted', color='yellow')
plt.plot(df['Date'], df['Resistance-2'], label='Resistance 2', linestyle='dotted', color='grey')
plt.plot(df['Date'], df['Support-2'], label='Support 2', linestyle='dotted', color='purple')
plt.plot(df['Date'], df['Resistance-3'], label='Resistance 3', linestyle='dotted', color='violet')
plt.plot(df['Date'], df['Support-3'], label='Support 3', linestyle='dotted', color='brown')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Nifty50 - Pivot Points and Price Levels')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
