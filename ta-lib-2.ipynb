```python
import yfinance as yf
import talib
import numpy as np

yf_data = yf.download("AAPL", start = "2014-01-05", end = "2014-07-04")
close_data = np.array(yf_data["Close"], dtype = "float64").flatten()

sma_30 = talib.SMA(close_data, timeperiod = 30)
rsi_20 = talib.RSI(close_data, timeperiod = 20)

print("SMA_30-day_data:",sma_30)
print("RSI_20-day_data:",rsi_20)
```

```python
import yfinance as yf
import talib
import numpy as np 

yf_data = yf.download("AAPL", start = "2023-01-05", end = "2024-01-04")
close_data = np.array(yf_data['Close'], dtype="float64").flatten()
length = len(close_data)

sma_20 = talib.SMA(close_data, timeperiod=20)
rsi_15 = talib.RSI(close_data, timeperiod=10)

signals = [0] * length

for i in range(length):  
    if close_data[i] > sma_20[i] and rsi_15[i] < 30:
        signals[i] = 1  
    elif close_data[i] < sma_20[i] and rsi_15[i] > 70:
        signals[i] = -1

print(signals)
print("SMA: ",sma_20[20:])
print("RSI: ",rsi_15[15:])
print("Data Length: ",length)
```

```python
import yfinance as yf
import talib
import numpy as np

yf_data = yf.download("MSFT", start = "2023-02-01", end = "2024-01-31")
close_data = np.array(yf_data["Close"], dtype = "float64").flatten()

sma_20 = talib.SMA(close_data, timeperiod = 20)
rsi_15 = talib.RSI(close_data, timeperiod = 15)

buy_signal = []
sell_signal = []

for i in range(1, len(close_data)):
    if close_data[i] > sma_20[i] and rsi_15[i] < 30:
        buy_signal.append(yf_data.index[i])
    elif close_data[i] < sma_20[i] and rsi_15[i] > 70:
        sell_signal.append(yf_data.index[i])
print("Buy Signal Data: ",buy_signal)
print("Sell Signal Data: ",sell_signal)
```

```python
import yfinance as yf
import talib
import numpy as np 

yf_data = yf.download("AAPL", start = "2023-01-05", end = "2024-01-04")
close_data = np.array(yf_data['Close'], dtype="float64").flatten()
length = len(close_data)

sma_20 = talib.SMA(close_data, timeperiod=20)

signals = [0] * length

for i in range(length):  
    if close_data[i] > sma_20[i] and close_data[i-1] < sma_20[i]:
        signals[i] = 1  
    elif close_data[i] < sma_20[i] and close_data[i] > sma_20[i]:
        signals[i] = -1

print(signals)
print("SMA: ",sma_20[:30])
print("Data Length: ",length)
```
