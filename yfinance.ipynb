# y-finance

```python
import yfinance as yf

apple= yf.Ticker("aapl")
print(apple)
```

```python
apple_historical_data = aapl.history(period = "1mo")
print(apple_historical_data)
```

```python
apple_historical_data = aapl.history(start = "2025-02-03", end = "2025-02-07", interval = "1m")
print(apple_historical_data)
```

```python
group_data = yf.download("AAPL GOOGL MSFT", start = "2024-01-01", end = "2024-06-30")
print(group_data)
```

```python
group_data = yf.download("AAPL GOOGL MSFT", start = "2024-01-01", end = "2024-06-30", group_by = "tickers")
print(group_data)
```
